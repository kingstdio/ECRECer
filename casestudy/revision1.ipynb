```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
data=pd.read_excel('/home/shizhenkun/codebase/DMLF/results/case2210/case_5121_integrated_results.xlsx')
testset = data.iloc[:, np.r_[0:2,3:5,7:12]]
testset.head(3)
```

```python
test_20 = testset[testset.ds_in.str.contains('_20') & (testset.ec_predictable) ].reset_index(drop=True)
test_22 = testset[testset.ds_in.str.contains('_22') & (testset.ec_predictable) ].reset_index(drop=True)
test_20.head(3)
```

```python
def get_first_digt(ecs):
    if ecs!='-':
        ecarray=ecs.split(',')

        res = [item.split('.')[0] for item in ecarray ]

        res = (',').join(list(set(res)))
        return res
    else:
        return '-'
    
def get_second_digt(ecs):
    if ecs!='-':
        ecarray=ecs.split(',')
        
        for item in ecarray:
            digitArray= item.split('.')
            res = [f'{digitArray[0]}.{digitArray[1]}']

        res = (',').join(list(set(res)))
        return res
    else:
        return '-'
    
def get_third_digt(ecs):
    if ecs!='-':
        ecarray=ecs.split(',')
        
        for item in ecarray:
            digitArray= item.split('.')
            res = [f'{digitArray[0]}.{digitArray[1]}.{digitArray[2]}']

        res = (',').join(list(set(res)))
        return res
    else:
        return '-'
```

```python
test_20.ec_number.apply(lambda x: get_third_digt(ecs=x))
```

```python
def caculateMetrix(groundtruth, predict, baselineName):
    acc = metrics.accuracy_score(groundtruth, predict)
    precision = metrics.precision_score(groundtruth, predict, average='macro', zero_division=True )
    recall = metrics.recall_score(groundtruth, predict, average='macro', zero_division=True)
    f1 = metrics.f1_score(groundtruth, predict, average='macro', zero_division=True)
    print('%12s'%baselineName, ' \t\t%f '%acc,'\t%f'% precision, '\t\t%f'% recall,'\t%f'% f1)
```

```python

methods =['ec_ecpred','ec_deepec','ec_catfam','ec_priam','ec_ours']
print('1 位 20')
print('%12s'%'baslineName', '\t\t', 'accuracy','\t', 'precision-macro \t', 'recall-macro','\t', 'f1-macro')
for method in methods:
    caculateMetrix(groundtruth=test_20.ec_number.apply(lambda x: get_first_digt(ecs=x)).to_list(), predict=test_20[method].apply(lambda x: get_first_digt(ecs=x)).to_list(), baselineName=method)

print('\n ', 15*'---------')
print('1 位 22')
print('%12s'%'baslineName', '\t\t', 'accuracy','\t', 'precision-macro \t', 'recall-macro','\t', 'f1-macro')
for method in methods:
    caculateMetrix(groundtruth=test_22.ec_number.apply(lambda x: get_first_digt(ecs=x)).to_list(), predict=test_22[method].apply(lambda x: get_first_digt(ecs=x)).to_list(), baselineName=method)
```
